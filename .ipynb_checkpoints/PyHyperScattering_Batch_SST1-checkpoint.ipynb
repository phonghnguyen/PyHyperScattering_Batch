```python
def get_immediate_subdirectories(a_dir):
    
    return [name for name in os.listdir(a_dir)
            if os.path.isdir(os.path.join(a_dir, name))]

def fmt(x, pos):
    a, b = '{:.0e}'.format(x).split('e')
    b = int(b)
    
    return r'${} \times 10^{{{}}}$'.format(a, b)

def find_edge(scan):
    if np.min(scan.energy) == 270:
        current_edge = 'C 1s'
        current_edge_folder = 'C1s'
    elif np.min(scan.energy) == 385:
        current_edge = 'N 1s'
        current_edge_folder = 'N1s'
    elif np.min(scan.energy) == 670:
        current_edge = 'F 1s'
        current_edge_folder = 'F1s'
    else:
        current_edge = 'unknown'
        current_edge_folder = 'unknown'
        
    return [current_edge, current_edge_folder]

def match_pols(scans):
    '''
    Iterates through list of scans in scans (made by appending scans to an initally empty list). 
    First scan in list is used as a reference and the corresponding scan of orthogonal polarization is found.
    As matches are found, items from the list are removed and returned, along with the scans that were matched.
    The intention is that the list is fed back to this function to find the next pair to process until the list is empty.
    '''
    # Case for finding 0 pol at list beginning, find matching 90 pol
    if scans[0].unstack('system').polarization == 0:
        current_scan_0pol = scans[0]
        del scans[0]
        [current_edge_0pol, current_edge_folder_0pol] = find_edge(current_scan_0pol)
        print(f'Scan {current_scan_0pol.sampleid} matches {current_scan_0pol.sample_name} {current_edge_0pol} ({current_scan_0pol.rsoxs_config}) with 0°  polarity')

        if len(scans) != 0:
            for i, scan in enumerate(scans):
                if scan.rsoxs_config == current_scan_0pol.rsoxs_config:
                    [current_edge_90pol, current_edge_folder_90pol] = find_edge(scan)
                    if current_edge_90pol == current_edge_0pol:
                        if scan.sample_name == current_scan_0pol.sample_name:
                            if scan.unstack('system').polarization == 90:
                                current_scan_90pol = scan
                                print(f'Scan {scan.sampleid} matches {scan.sample_name} {current_edge_90pol} ({current_scan_90pol.rsoxs_config}) with 90° polarity')
                                del scans[i]
                                return [current_scan_0pol, current_scan_90pol, scans]
            current_scan_90pol = []
        else:
            print('Reached end, unable to pair last scan')
            
    # Case for finding 90 pol at list beginning, find matching 0 pol
    if scans[0].unstack('system').polarization == 90:
        current_scan_90pol = scans[0]
        del scans[0]
        
        [current_edge_90pol, current_edge_folder_90pol] = find_edge(current_scan_90pol)
            
        print(f'Scan {current_scan_90pol.sampleid} matches {current_scan_90pol.sample_name} {current_edge_90pol} ({current_scan_90pol.rsoxs_config}) with 90° polarity')

        if len(scans) != 0:
            for i, scan in enumerate(scans):
                if scan.rsoxs_config == current_scan_90pol.rsoxs_config:
                    [current_edge_0pol, current_edge_folder_0pol] = find_edge(scan)
                    if current_edge_0pol == current_edge_90pol:   
                        if scan.sample_name == current_scan_90pol.sample_name:
                            if scan.unstack('system').polarization == 0:
                                current_scan_0pol = scan
                                [current_edge, current_edge_folder] = find_edge(current_scan_0pol)
                                print(f'Scan {scan.sampleid} matches {scan.sample_name} {current_edge_0pol} ({current_scan_0pol.rsoxs_config}) with 0°  polarity')
                                del scans[i]
                                return [current_scan_0pol, current_scan_90pol, scans]
            current_scan_0pol = []
        else:
            print('Reached end, unable to pair last scan')
            
    return [current_scan_0pol, current_scan_90pol, scans]

def assign_mask(scan):
    if scan.rsoxs_config == 'waxs':
        mask_method = 'nika'
        mask = 'SST1_Sept2022_WAXS_mask.hdf'
    elif scan.rsoxs_config == 'saxs':
        mask_method = 'none'
        mask = []
    else:
        mask_method = 'none'
        mask = []
    
    return [mask_method, mask]

def make_output_dir(base_save_dir, scan, tail):
# tail is a string that describes the data file being exported, used as folder prefix
    [current_edge, current_edge_folder] = find_edge(scan)
    
    # create directories for output files
    save_dir = os.path.join(base_save_dir, f'{tail}_{sample_name_root(scan)}')
    os.makedirs(save_dir, exist_ok = True)
    
    return save_dir

def clear_dir(dir):
    for file_name in os.listdir(dir):
        file_path = os.path.join(dir, file_name)
        try:
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)
        except Exception as e:
            print('Failed to delete %s. Reason: %s' % (file_path, e))
            
def sample_name_root(scan):
    [current_edge, current_edge_folder] = find_edge(scan)
    return f'{current_edge_folder}_{scan.sample_name}'

def scan_loader_SST1(base_data_dirs, exclude = [], ll = [], ul = []):
    '''
    Args:
        base_data_dirs (list of str): path to the directories to load
        exlude (list of str): list of scan IDs to exclude from loading into scans
        ll (str): scan ID below which to exclude from loading into scans
        ul (str) scan ID above which to exclude from loading into scans
    '''
    base_data_dir_tracker = 0
    scan_tracker = 0
    scan_skip_corr = 0
    scans = []
    for base_data_dir in base_data_dirs:
        base_data_dir_tracker += 1
        scan_tracker = 0
        scan_skip_corr = 0
        all_scans = pd.Series(get_immediate_subdirectories(base_data_dir))
        for current_scan in np.sort(all_scans):
            scan_tracker += 1
            if ((len(ll) == 0) and (len(ul) == 0)):
                if current_scan in exclude:
                    print(f'Skipping scan {current_scan} - ID in list to exclude')
                    scan_skip_corr += 1
                else:
                    clear_output(wait = True)
                    print(f'RAM usage: {psutil.virtual_memory().available * 100 / psutil.virtual_memory().total :.2f} %')
                    print(f'CPU usage: {psutil.cpu_percent() :.2f} %')
                    print(f'Currently pulling data from \'{base_data_dir}\' ({base_data_dir_tracker}/{len(base_data_dirs)})')
                    print(f'Currently loading scan {current_scan} ({scan_tracker - scan_skip_corr}/{len(all_scans) - scan_skip_corr})')
                    scan = file_loader.loadFileSeries(base_data_dir + current_scan, ['energy', 'polarization'])
                    scans.append(scan)
            elif ((len(ll) > 0) and (len(ul) == 0)):
                if current_scan >= ll:
                    if current_scan in exclude:
                        print(f'Skipping scan {current_scan} - ID in list to exclude')
                        scan_skip_corr += 1
                    else:
                        clear_output(wait = True)
                        print(f'RAM usage: {psutil.virtual_memory().available * 100 / psutil.virtual_memory().total :.2f} %')
                        print(f'CPU usage: {psutil.cpu_percent() :.2f} %')
                        print(f'Currently pulling data from \'{base_data_dir}\' ({base_data_dir_tracker}/{len(base_data_dirs)})')
                        print(f'Currently loading scan {current_scan} ({scan_tracker - scan_skip_corr}/{len(all_scans) - scan_skip_corr})')
                        scan = file_loader.loadFileSeries(base_data_dir + current_scan, ['energy', 'polarization'])
                        scans.append(scan)
            elif ((len(ll) == 0) and (len(ul) > 0)):
                if current_scan <= ul:
                    if current_scan in exclude:
                        print(f'Skipping scan {current_scan} - ID in list to exclude')
                        scan_skip_corr += 1
                    else:
                        clear_output(wait = True)
                        print(f'RAM usage: {psutil.virtual_memory().available * 100 / psutil.virtual_memory().total :.2f} %')
                        print(f'CPU usage: {psutil.cpu_percent() :.2f} %')
                        print(f'Currently pulling data from \'{base_data_dir}\' ({base_data_dir_tracker}/{len(base_data_dirs)})')
                        print(f'Currently loading scan {current_scan} ({scan_tracker - scan_skip_corr}/{len(all_scans) - scan_skip_corr})')
                        scan = file_loader.loadFileSeries(base_data_dir + current_scan, ['energy', 'polarization'])
                        scans.append(scan)
            elif ((len(ll) > 0) and (len(ul) > 0)):
                if ((current_scan >= ll) and (current_scan <= ul)):
                    if current_scan in exclude:
                        print(f'Skipping scan {current_scan} - ID in list to exclude')
                        scan_skip_corr += 1
                    else:
                        clear_output(wait = True)
                        print(f'RAM usage: {psutil.virtual_memory().available * 100 / psutil.virtual_memory().total :.2f} %')
                        print(f'CPU usage: {psutil.cpu_percent() :.2f} %')
                        print(f'Currently pulling data from \'{base_data_dir}\' ({base_data_dir_tracker}/{len(base_data_dirs)})')
                        print(f'Currently loading scan {current_scan} ({scan_tracker - scan_skip_corr}/{len(all_scans) - scan_skip_corr})')
                        scan = file_loader.loadFileSeries(base_data_dir + current_scan, ['energy', 'polarization'])
                        scans.append(scan)

    return scans

def setup_plot():
    plt.rcParams.update({"font.size": 18})
    plt.rcParams.update({"axes.linewidth": 2})

    fig, axes = plt.subplots(1, 1, figsize = (3.5, 3.5))

    # Edit the major and minor ticks of the x and y axes
    axes.xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
    axes.xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
    axes.yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
    axes.yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')
    
    return [fig, axes]
    
def save_fig(fig, scan, current_energy, save_dir, tail, res):
    if current_energy > 0:
        save_path = os.path.join(save_dir, f'{tail}_{sample_name_root(scan)}_{current_energy}eV.png')
        fig.savefig(save_path, bbox_inches = 'tight', pad_inches = 0.2, dpi = res, transparent = False)
    else:
        save_path = os.path.join(save_dir, f'{tail}_{sample_name_root(scan)}.png')
        fig.savefig(save_path, bbox_inches = 'tight', pad_inches = 0.2, dpi = res, transparent = False)
        
def make_gif(scan, input_dir, output_dir, tail):
    image_folder = os.fsencode(input_dir)

    file_names = []

    for file in os.listdir(image_folder):
        file_name = os.fsdecode(file)
        file_names.append(file_name)

    file_names.sort() # this iteration technique has no built in order, so sort the frames

    # Make it pause at the end so that the viewers can ponder
    for _ in range(10):
        file_names.append(file_names[-1])

    images = list(map(lambda file_name: imageio.imread(os.path.join(input_dir, file_name)), file_names))

    imageio.mimsave(os.path.join(output_dir, f'{tail}_{sample_name_root(scan)}.gif'), images, duration = 0.5) # modify duration as needed
```

```python
print(f'{len(scans)} scans imported')

# variables to append sample ID and AR xarrays
sample_names = []
edges = []

ARs = []
AR_scales = []

paras = []
para_scales = []

perps = []
perp_scales = []

circs = []
circ_scales = []

while len(scans) > 0:
    
    # reset energy
    current_energy = 0
        
    # Find samples with matching experimental configurations, elemental edges, names, and corresponding polarizations
    # also takes care of removing already processed scans from the list of scans remaining
    current_scan_pol0 = []
    current_scan_pol90 = []
    
    while ((len(current_scan_pol0) == 0) or (len(current_scan_pol90) == 0) and (len(scans) > 0)):
        [current_scan_pol0, current_scan_pol90, scans] = match_pols(scans)
    
    # Find the elemental edge of the current scan
    [current_edge, current_edge_folder] = find_edge(current_scan_pol0)
    
    # log info for mosaics
    sample_names.append(current_scan_pol0.sample_name)
    edges.append(current_edge)
    
    # Set the energy and q range of interest depending on the elemental edge of the current scan
    if current_edge == 'C 1s':
        start_en = 280
        end_en = 295
        start_q = 0.01
        end_q = 0.1
    else:
        start_en = np.nanmin(current_scan_pol0.energy)
        end_en = np.nanmax(current_scan_pol0.energy)
        start_q = 0.01
        end_q = np.nanmax(current_scan_pol0.q)
        
    # Assign the mask depending on experimental config
    [mask_method, mask] = assign_mask(current_scan_pol0)
    
    # import mask into PyHyperScattering integrator object
    integ.loadNikaMask(filetoload = mask, rotate_image = False)

    # rotate mask to match experimental orientation
    integ.mask = np.flipud(np.rot90(integ.mask))

    # compare experimental and mask orientation
#     f, axarr = plt.subplots(1,2)
#     axarr[0].imshow(scan_unstacked.sel(energy=np.max(scan.energy)),
#                     # norm = LogNorm(1,np.nanpercentile(scan_unstacked.sel(energy=current_energy),99)
#                     vmin = 0,
#                     vmax = np.nanpercentile(scan_unstacked.sel(energy=np.max(scan.energy)), 90),
#                     cmap = 'viridis',
#                     aspect='1')
#     axarr[1].imshow(integ.mask, norm = LogNorm(1,np.nanpercentile(scan_unstacked.sel(energy=np.max(scan.energy)),90)))
    
    # make sure final processing output directory exists
    reduced_save_dir = make_output_dir(base_save_dir, current_scan_pol0, '0_reduced')
    
    # check whether sample has been fully processed
    save_path = os.path.join(reduced_save_dir, f'abs_linecuts_{sample_name_root(current_scan_pol0)}.gif')
    tail = os.path.split(save_path)[1]

    if tail not in os.listdir(reduced_save_dir):
    # if True:
    
        # Make sure directories exist
        rel_detImage_save_dir = make_output_dir(base_save_dir, current_scan_pol0, 'rel_detImage')
        abs_detImage_save_dir = make_output_dir(base_save_dir, current_scan_pol0, 'abs_detImage')

        rel_linecuts_save_dir = make_output_dir(base_save_dir, current_scan_pol0, 'rel_linecuts')
        abs_linecuts_save_dir = make_output_dir(base_save_dir, current_scan_pol0, 'abs_linecuts')
    
        # Remove old files
        clear_dir(rel_detImage_save_dir)
        clear_dir(abs_detImage_save_dir)

        clear_dir(rel_linecuts_save_dir)
        clear_dir(abs_linecuts_save_dir)

        clear_dir(reduced_save_dir)
    
        print(f'RAM usage: {psutil.virtual_memory().available * 100 / psutil.virtual_memory().total :.2f} %')
        print(f'CPU usage: {psutil.cpu_percent() :.2f} %')
        
        print(f'Integrating scan {current_scan_pol0.sampleid}:')
              
        integrated_data_pol0 = integ.integrateImageStack(current_scan_pol0)
        integrated_data_pol0_unstacked = integrated_data_pol0.unstack('system')
        
        print(f'RAM usage: {psutil.virtual_memory().available * 100 / psutil.virtual_memory().total :.2f} %')
        print(f'CPU usage: {psutil.cpu_percent() :.2f} %')
        
        print(f'Integrating scan {current_scan_pol90.sampleid}:')
        
        integrated_data_pol90 = integ.integrateImageStack(current_scan_pol90)
        integrated_data_pol90_unstacked = integrated_data_pol90.unstack('system')
        
        # calculate total intensities to normalize by
        total_intensity_pol0 = integrated_data_pol0_unstacked.sel(q = slice(start_q, end_q,), polarization = 0).mean('q').mean('energy').mean('chi')
        total_intensity_pol90 = integrated_data_pol90_unstacked.sel(q = slice(start_q, end_q,), polarization = 90).mean('q').mean('energy').mean('chi')
        
        total_intensity_pol0_corr = (total_intensity_pol0 + total_intensity_pol90)/2/total_intensity_pol0
        total_intensity_pol90_corr = (total_intensity_pol0 + total_intensity_pol90)/2/total_intensity_pol90
        
        clear_output(wait = True)
        
        
        
        
        
        
        
        
        
        
        
        
        
           
        ## Make plot of the AR
        pol0_para = integrated_data_pol0_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(q = slice(start_q, end_q,), polarization = 0)*total_intensity_pol0_corr
        pol0_perp = integrated_data_pol0_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr

        pol90_para = integrated_data_pol90_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr
        pol90_perp = integrated_data_pol90_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr

        para = pol0_para + pol90_para
        perp = pol0_perp + pol90_perp

        AR_pol0 = (pol0_para - pol0_perp)/(pol0_para + pol0_para)
        AR_pol90 = (pol90_para - pol90_perp)/(pol90_para + pol90_para)
        
        AR = (AR_pol0 + AR_pol90)/2

        scale = np.nanmax([abs(np.nanpercentile(AR.sel(energy = slice(280, 295)), 1)), abs(np.nanpercentile(AR.sel(energy = slice(start_en, end_en)), 99))])
        # scale = np.nanmax([abs(np.nanmin(AR.sel(energy = slice(280, 295)))), abs(np.nanmax(AR.sel(energy = slice(start_en, end_en))))])
        
        AR_scales.append(scale)
    
        # log info for mosaic
        ARs.append(AR)    

        [fig, axes] = setup_plot()

        im_cbar = AR.sel(energy = slice(start_en, end_en)).plot(
            # norm = LogNorm(1,current_vmax_cake[-1]),
            x = 'q', y = 'energy',
            ax = axes,
            vmin = -scale,
            vmax = scale,
            cmap = 'RdBu',
            add_colorbar = False)

        cbar_ax = fig.add_axes([0.975, 0.145, 0.03, 0.715])
        cbar = fig.colorbar(im_cbar, cax = cbar_ax)

        # plt.setp(cbar.ax.get_yticklabels(), ha = "right")
        # cbar.ax.tick_params(pad = 37.5)
        cbar.ax.set_ylabel('Anisotropy')

        fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 20, y = 1, x = 0.5)
        axes.set_xlabel('$\it{q}$ (Å$^{-1}$)')
        axes.set_ylabel('Energy (eV)')
        axes.set_xlim([start_q, end_q])
        # axes.set_xticks([0.01, 0.03, 0.05, 0.07, 0.09], ['', '0.03', '0.05', '0.07', '0.09'])
        axes.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))
        axes.set_ylim([start_en, end_en])

        axes.xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
        axes.xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
        axes.yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
        axes.yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

        plt.show()
        
        save_fig(fig, current_scan_pol0, current_energy, reduced_save_dir, 'AR', 450)


        [fig, axes] = setup_plot()

        im_cbar = para.plot(
            # norm = LogNorm(1,current_vmax_cake[-1]),
            x = 'q', y = 'energy',
            ax = axes,
            vmin = np.nanpercentile(para, 97.5),
            vmax = scale,
            cmap = cmap,
            add_colorbar = False)
        
        paras.append(para)
        para_scales.append(np.nanpercentile(para, 97.5))

        cbar_ax = fig.add_axes([0.975, 0.145, 0.03, 0.715])
        fig.colorbar(im_cbar, cax = cbar_ax)

        fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 20, y = 1, x = 0.5)
        axes.set_title('')
        axes.text(0.975, 0.91, f'Parallel', color = 'white', fontsize = 15, transform = axes.transAxes, horizontalalignment = 'right')

        axes.set_xlabel('$\it{q}$ (Å$^{-1}$)')
        axes.set_ylabel('Energy (eV)')
        axes.set_xlim([start_q, end_q])
        # axes.set_xticks([0.01, 0.03, 0.05, 0.07, 0.09], ['', '0.03', '0.05', '0.07', '0.09'])
        axes.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))
        axes.set_ylim([start_en, end_en])

        axes.xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
        axes.xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
        axes.yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
        axes.yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

        plt.show()
        
        save_fig(fig, current_scan_pol0, current_energy, reduced_save_dir, 'para', 450)


        [fig, axes] = setup_plot()

        im_cbar = perp.plot(
            # norm = LogNorm(1,current_vmax_cake[-1]),
            x = 'q', y = 'energy',
            ax = axes,
            vmin = np.nanpercentile(perp, 97.5),
            vmax = scale,
            cmap = cmap,
            add_colorbar = False)
        
        perps.append(perp)
        perp_scales.append(np.nanpercentile(perp, 97.5))

        cbar_ax = fig.add_axes([0.975, 0.145, 0.03, 0.715])
        fig.colorbar(im_cbar, cax = cbar_ax)

        fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 20, y = 1, x = 0.5)
        axes.set_title('')
        axes.text(0.975, 0.91, f'Perpendicular', color = 'white', fontsize = 15, transform = axes.transAxes, horizontalalignment = 'right')

        axes.set_xlabel('$\it{q}$ (Å$^{-1}$)')
        axes.set_ylabel('Energy (eV)')
        axes.set_xlim([start_q, end_q])
        # axes.set_xticks([0.01, 0.03, 0.05, 0.07, 0.09], ['', '0.03', '0.05', '0.07', '0.09'])
        axes.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))
        axes.set_ylim([start_en, end_en])

        axes.xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
        axes.xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
        axes.yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
        axes.yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

        plt.show()
        
        save_fig(fig, current_scan_pol0, current_energy, reduced_save_dir, 'perp', 450)


        para_perp = para + perp

        [fig, axes] = setup_plot()

        im_cbar = para_perp.plot(
            # norm = LogNorm(1,current_vmax_cake[-1]),
            x = 'q', y = 'energy',
            ax = axes,
            vmin = np.nanpercentile(para_perp, 97.5),
            vmax = scale,
            cmap = cmap,
            add_colorbar = False)
        
        circs.append(para_perp)
        circ_scales.append(np.nanpercentile(para_perp, 97.5))
        
        cbar_ax = fig.add_axes([0.975, 0.145, 0.03, 0.715])
        fig.colorbar(im_cbar, cax = cbar_ax)

        fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 20, y = 1, x = 0.5)
        axes.set_title('')
        axes.text(0.975, 0.91, f'Circular', color = 'white', fontsize = 15, transform = axes.transAxes, horizontalalignment = 'right')

        axes.set_xlabel('$\it{q}$ (Å$^{-1}$)')
        axes.set_ylabel('Energy (eV)')
        axes.set_xlim([start_q, end_q])
        # axes.set_xticks([0.01, 0.03, 0.05, 0.07, 0.09], ['', '0.03', '0.05', '0.07', '0.09'])
        axes.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))
        axes.set_ylim([start_en, end_en])

        axes.xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
        axes.xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
        axes.yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
        axes.yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

        plt.show()
        
        save_fig(fig, current_scan_pol0, current_energy, reduced_save_dir, 'circ', 450)


        integrated_data_unstacked = (integrated_data_pol0_unstacked.sel(polarization = 0)*total_intensity_pol0_corr) + (integrated_data_pol90_unstacked.sel(polarization = 90)*total_intensity_pol90_corr)

        [fig, axes] = setup_plot()

        FY_NEXAFS = integrated_data_unstacked.mean('chi').sel(q = slice(0.055, 0.075), energy = slice(start_en, end_en)).mean('q')
        
        FY_NEXAFS.plot(
            ax = axes,
            color = 'k')

        fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 20, y = 1, x = 0.5)
        axes.set_ylabel('Intensity (a.u.)')
        axes.set_xlabel('Energy (eV)')
        axes.set_xlim([start_en, end_en])
        axes.set_ylim([0.9*np.nanmin(FY_NEXAFS), 
                       1.1*np.nanmax(FY_NEXAFS)])

        plt.show()
        
        save_fig(fig, current_scan_pol0, current_energy, reduced_save_dir, 'FY_NEXAFS', 450)


        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        # variables to hist energy-dependent contrasts and plot limits
        current_vmax_detImage = []
        current_linecuts_ylim_min = []
        current_linecuts_ylim_max = []
        
        ### Loop for iterating through energies
        for current_energy in np.sort(integrated_data_pol0_unstacked.sel(energy = slice(start_en, end_en)).energy):

            ## Make plots of detector image            
            current_scan_pol0_unstacked = current_scan_pol0.unstack('system')
            current_scan_pol90_unstacked = current_scan_pol90.unstack('system')

            current_vmax_detImage.append(np.min([np.nanpercentile(current_scan_pol0_unstacked.sel(energy = current_energy)*total_intensity_pol0_corr, 97.5), 
                                        np.nanpercentile(current_scan_pol90_unstacked.sel(energy = current_energy)*total_intensity_pol90_corr, 97.5)]))

            current_vmax_cake = current_vmax_detImage

            fig, axes = plt.subplots(2, 2, figsize = (7, 7))

            # im_pol0 = axes[0, 0].imshow(current_scan_pol0_unstacked.sel(energy = current_energy, polarization = 0).where(~np.array(integ.mask)),
            im_pol0 = axes[0, 0].imshow(current_scan_pol0_unstacked.sel(energy = current_energy, polarization = 0)*total_intensity_pol0_corr,
                            # norm = LogNorm(1, np.nanmax(current_vmax_detImage)),
                            vmin = 0,
                            vmax = current_vmax_detImage[-1],
                            cmap = cmap,
                            aspect = 1)

            axes[0, 0].set_xticks([])
            axes[0, 0].set_yticks([])

            # im_pol90 = axes[0, 1].imshow(current_scan_pol90_unstacked.sel(energy = current_energy, polarization = 90).where(~np.array(integ.mask)),
            im_pol90 = axes[0, 1].imshow(current_scan_pol90_unstacked.sel(energy = current_energy, polarization = 90)*total_intensity_pol90_corr,
                            # norm = LogNorm(1, np.nanmax(current_vmax_detImage)),
                            vmin = 0,
                            vmax = current_vmax_detImage[-1],
                            cmap = cmap,
                            aspect = 1)

            axes[0, 1].set_xticks([])
            axes[0, 1].set_yticks([])

            cake_pol0 = integrated_data_pol0_unstacked.sel(energy = current_energy, q = slice(start_q, end_q))*total_intensity_pol0_corr
            
            im = cake_pol0.plot(
                x = 'chi', y = 'q',
                ax = axes[1, 0],
                # norm = LogNorm(1, current_vmax_detImage[-1]),
                vmin = 0,
                vmax = current_vmax_detImage[-1],
                cmap = cmap,
                add_colorbar = False)

            axes[1, 0].set_ylabel('$\it{q}$ (Å$^{-1}$)')
            axes[1, 0].set_xlabel('Azimuth, $\it{χ}$')
            axes[1, 0].set_yticks([0.01, 0.03, 0.05, 0.07, 0.09])
            axes[1, 0].set_xticks([-180, -90, 0, 90, 180], ['', '-90°', '0°', '90°', ''])
            # axes[1, 0].set_xticks([-180, -90, 0, 90, 180], ['-π', '', '0', '', 'π'])

            axes[1, 0].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
            axes[1, 0].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
            axes[1, 0].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', right = 'on')
            axes[1, 0].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', right = 'on')
            
            
            cake_pol90 = integrated_data_pol90_unstacked.sel(energy = current_energy, q = slice(start_q, end_q))*total_intensity_pol90_corr

            im = cake_pol90.plot(
                x = 'chi', y = 'q',
                ax = axes[1, 1],
                # norm = LogNorm(1, current_vmax_detImage[-1]),
                vmin = 0,
                vmax = current_vmax_detImage[-1],
                cmap = cmap,
                add_colorbar = False)

            axes[1, 1].set_xlabel('Azimuth, $\it{χ}$')
            axes[1, 1].set_ylabel('')
            axes[1, 1].set_xticks([-180, -90, 0, 90, 180], ['', '-90°', '0°', '90°', '180°'])
            # axes[1, 1].set_xticks([-180, -90, 0, 90, 180], ['-π', '', '0', '', 'π'])
            axes[1, 1].set_yticks([0.01, 0.03, 0.05, 0.07, 0.09])
            axes[1, 1].set_yticklabels([])

            axes[1, 1].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
            axes[1, 1].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
            axes[1, 1].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', right = 'on')
            axes[1, 1].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', right = 'on')

            cbar_ax = fig.add_axes([0.975, 0.145, 0.02, 0.725])
            cbar = fig.colorbar(im_pol90, cax = cbar_ax)
            # cbar.ax.set_ylabel('Intensity (a.u.)')

            fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 17.5, y = 0.95, x = 0.55)
            axes[0, 0].set_title('')
            axes[0, 1].set_title('')
            axes[1, 0].set_title('')
            axes[1, 1].set_title('')

            axes[0, 0].text(0.975, 0.91, f'{current_energy:.2f} eV, 0°', color = 'white', fontsize = 15, transform = axes[0, 0].transAxes, horizontalalignment = 'right')
            axes[0, 1].text(0.975, 0.91, f'{current_energy:.2f} eV, 90°', color = 'white', fontsize = 15, transform = axes[0, 1].transAxes, horizontalalignment = 'right')
            axes[1, 0].text(0.975, 0.91, f'{current_energy:.2f} eV, 0°', color = 'white', fontsize = 15, transform = axes[1, 0].transAxes, horizontalalignment = 'right')
            axes[1, 1].text(0.975, 0.91, f'{current_energy:.2f} eV, 90°', color = 'white', fontsize = 15, transform = axes[1, 1].transAxes, horizontalalignment = 'right')

            plt.tight_layout()

            # plt.show()

            save_fig(fig, current_scan_pol0, current_energy, rel_detImage_save_dir, 'detImage', 450)





            [fig, axes] = plt.subplots(2, 2, figsize = (7, 7))
            fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 20, y = 0.975, x = 0.5)
            axes[1, 1].axis('off')

            pol0_para = integrated_data_pol0_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = current_energy, q = slice(start_q, end_q,), polarization = 0)*total_intensity_pol0_corr
            pol0_perp = integrated_data_pol0_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = current_energy, q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr

            pol90_para = integrated_data_pol90_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = current_energy, q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr
            pol90_perp = integrated_data_pol90_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = current_energy, q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr

            para = pol0_para + pol90_para
            perp = pol0_perp + pol90_perp

            axes[0, 0].errorbar(np.asarray(para.q).flatten(), np.asarray(para).flatten(), np.sqrt(para.values).flatten(),
                        linestyle ='none', color = (57/255, 115/255, 180/255), alpha=0.25)

            axes[0, 0].plot(para.q, para, '-', color = (57/255, 115/255, 180/255), label = 'Parallel')

            axes[0, 0].errorbar(np.asarray(perp.q).flatten(), np.asarray(perp).flatten(), np.sqrt(perp.values).flatten(),
                        linestyle ='none', color = (170/255, 21/255, 40/255), alpha=0.25)

            axes[0, 0].plot(perp.q, perp, '-', color = (170/255, 21/255, 40/255), label = 'Perpendicular')

            axes[0, 0].legend(loc = 'upper right', frameon = False, fontsize = 15)

            axes[0, 0].set_title('')
            axes[0, 0].text(0.395, 0.6, f'{current_energy:.2f} eV', color = 'black', fontsize = 15, transform = axes[0, 0].transAxes, horizontalalignment = 'left')

            axes[0, 0].set_ylabel('Intensity (a.u.)')
            axes[0, 0].set_xlabel('')
            axes[0, 0].set_xlim([start_q, end_q])
            axes[0, 0].set_ylim([np.nanmin([np.nanpercentile(para, 1), np.nanpercentile(perp, 1)]),
                     np.nanmax([np.nanpercentile(para, 99.9), np.nanpercentile(perp, 99.9)])])
            axes[0, 0].set_yticklabels([])
            axes[0, 0].set_xticklabels([])
            axes[0, 0].set_xticks([0.01, 0.03, 0.05, 0.07, 0.09])

            axes[0, 0].yaxis.set_major_formatter(ticker.FuncFormatter(fmt))
            axes[0, 0].set_yscale('log')

            axes[0, 0].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', top = 'on')
            axes[0, 0].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', top = 'on')
            axes[0, 0].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
            axes[0, 0].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

            current_linecuts_ylim_min.append(np.nanmin(axes[0, 0].get_ylim()))
            current_linecuts_ylim_max.append(np.nanmax(axes[0, 0].get_ylim()))


            FY_NEXAFS.plot(
                ax = axes[0, 1],
                color = 'k',
                alpha = 0.2)

            integrated_data_unstacked.mean('chi').sel(q = slice(0.055, 0.075), energy = slice(start_en, current_energy)).mean('q').plot(
                ax = axes[0, 1],
                color = 'k',
                alpha = 1)

            axes[0, 1].plot(current_energy, integrated_data_unstacked.mean('chi').sel(q = slice(0.055, 0.075)).mean('q').sel(energy = current_energy), 'o', color = 'black')

            axes[0, 1].set_title('Fluorescence NEXAFS', color = 'black', fontsize = 18)
            # axes[0, 1].text(0.975, 0.05, f'FY NEXAFS', color = 'black', fontsize = 15, transform = axes[0, 1].transAxes, horizontalalignment = 'right')

            # axes[0, 1].set_ylabel('Intensity (a.u.)')
            axes[0, 1].set_xlabel('Energy (eV)')
            axes[0, 1].set_xlim([start_en, end_en])
            axes[0, 1].set_ylim([0.9*np.nanmin(FY_NEXAFS), 
                                 1.1*np.nanmax(FY_NEXAFS)])
            axes[0, 1].set_yticklabels([])
            # axes[0, 1].set_yscale('log')

            axes[0, 1].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
            axes[0, 1].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
            axes[0, 1].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
            axes[0, 1].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')


            pol0_para = integrated_data_pol0_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = slice(start_en, current_energy), q = slice(start_q, end_q,), polarization = 0)*total_intensity_pol0_corr
            pol0_perp = integrated_data_pol0_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = slice(start_en, current_energy), q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr

            pol90_para = integrated_data_pol90_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = slice(start_en, current_energy), q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr
            pol90_perp = integrated_data_pol90_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = slice(start_en, current_energy), q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr

            AR_pol0 = (pol0_para - pol0_perp)/(pol0_para + pol0_para)
            AR_pol90 = (pol90_para - pol90_perp)/(pol90_para + pol90_para)

            AR = (AR_pol0 + AR_pol90)/2
            
            if len(AR.energy) < 2:
                pol0_para = integrated_data_pol0_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = slice(start_en, end_en), q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr
                pol0_perp = integrated_data_pol0_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = slice(start_en, end_en), q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr

                pol90_para = integrated_data_pol90_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = slice(start_en, end_en), q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr
                pol90_perp = integrated_data_pol90_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = slice(start_en, end_en), q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr

                AR_pol0 = (pol0_para - pol0_perp)/(pol0_para + pol0_para)
                AR_pol90 = (pol90_para - pol90_perp)/(pol90_para + pol90_para)

                AR1 = (AR_pol0 + AR_pol90)/2

                scale = np.nanmax([abs(np.nanpercentile(AR1.sel(energy = slice(start_en, end_en)), 1)), abs(np.nanpercentile(AR1.sel(energy = slice(start_en, end_en)), 99))])
                # scale = np.nanmax([abs(np.nanmin(AR.sel(energy = slice(280, 295)))), abs(np.nanmax(AR.sel(energy = slice(280, 295))))])

                im_cbar = AR1.plot(
                    x = 'q', y = 'energy',
                    ax = axes[1, 0],
                    vmin = -scale,
                    vmax = scale,
                    cmap = 'RdBu',
                    add_colorbar = False)

                cbar_ax = fig.add_axes([0.5, 0.14, 0.02, 0.31])
                cbar = fig.colorbar(im_cbar, cax = cbar_ax)

                # plt.setp(cbar.ax.get_yticklabels(), ha = "right")
                # cbar.ax.tick_params(pad = 50)
                # cbar.ax.set_ylabel('Anisotropy')

                axes[1, 0].set_title('')
                # axes[1, 0].text(0.975, 0.05, f'AR', color = 'black', fontsize = 15, transform = axes[1, 0].transAxes, horizontalalignment = 'right')

                axes[1, 0].set_xlabel('$\it{q}$ (Å$^{-1}$)')
                axes[1, 0].set_ylabel('Energy (eV)')
                axes[1, 0].set_xlim([start_q, end_q])
                axes[1, 0].set_xticks([0.01, 0.03, 0.05, 0.07, 0.09], ['', '0.03', '0.05', '0.07', '0.09'])
                axes[1, 0].set_ylim([AR1.energy[0], (AR1.energy[1] - AR1.energy[0])/2 + AR1.energy[0]])
                # axes[1, 0].set_ylim([np.nanmin(AR1.energy), np.nanmax(AR1.energy)])
                axes[1, 0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

                # ticks = [tick for tick in axes[1, 0].get_yticks()]
                # del ticks[-1]
                # axes[1, 0].set_yticks(ticks)

                axes[1, 0].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', top = 'on')
                axes[1, 0].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', top = 'on')
                axes[1, 0].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
                axes[1, 0].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

            else:

                AR1.plot(
                    x = 'q', y = 'energy',
                    ax = axes[1, 0],
                    vmin = -scale,
                    vmax = scale,
                    cmap = 'RdBu',
                    add_colorbar = False,
                    alpha = 0.2)

                im_cbar = AR.plot(
                    x = 'q', y = 'energy',
                    ax = axes[1, 0],
                    vmin = -scale,
                    vmax = scale,
                    cmap = 'RdBu',
                    add_colorbar = False)

                cbar_ax = fig.add_axes([0.5, 0.14, 0.02, 0.31])
                cbar = fig.colorbar(im_cbar, cax = cbar_ax)

                # plt.setp(cbar.ax.get_yticklabels(), ha = "right")
                # cbar.ax.tick_params(pad = 50)
                cbar.ax.set_ylabel('Anisotropy')

                axes[1, 0].set_title('')
                # axes[1, 0].text(0.975, 0.05, f'AR', color = 'black', fontsize = 15, transform = axes[1, 0].transAxes, horizontalalignment = 'right')

                axes[1, 0].set_xlabel('$\it{q}$ (Å$^{-1}$)')
                axes[1, 0].set_ylabel('Energy (eV)')
                axes[1, 0].set_xlim([start_q, end_q])
                axes[1, 0].set_xticks([0.01, 0.03, 0.05, 0.07, 0.09], ['', '0.03', '0.05', '0.07', '0.09'])

                # ticks = [tick for tick in axes[1, 0].get_yticks()]
                # del ticks[-1]
                # axes[1, 0].set_yticks(ticks)
                axes[1, 0].set_ylim([np.nanmin(AR.energy), np.nanmax(AR.energy)])
                axes[1, 0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

                axes[1, 0].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', top = 'on')
                axes[1, 0].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', top = 'on')
                axes[1, 0].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
                axes[1, 0].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

            clear_output(wait = True)

            # plt.tight_layout()

            plt.show()
            
            save_fig(fig, current_scan_pol0, current_energy, rel_linecuts_save_dir, 'linecuts', 450)
            
            print(f'{len(scans)} scans remaining')
            
        clear_output(wait = True)    
        
        print(f'{len(scans)} scans remaining')
        
        print('Making gif of rel detector images')    
        make_gif(current_scan_pol0, rel_detImage_save_dir, reduced_save_dir, 'rel_detImage')
        
        print('Making gif of rel linecuts images')    
        make_gif(current_scan_pol0, rel_linecuts_save_dir, reduced_save_dir, 'rel_linecuts')
        
        print('rel gifs done, purging temporary images')  
        shutil.rmtree(rel_detImage_save_dir)
        shutil.rmtree(rel_linecuts_save_dir)
 
        
               
        
        
        
        
        
        
        
        
        
        


        ### Loop for iterating through energies
        for current_energy in np.sort(integrated_data_pol0_unstacked.sel(energy = slice(start_en, end_en)).energy):

            # ## Make plots of detector image            
            # current_scan_pol0_unstacked = current_scan_pol0.unstack('system')
            # current_scan_pol90_unstacked = current_scan_pol90.unstack('system')

            # current_vmax_detImage.append(np.min([np.nanpercentile(current_scan_pol0_unstacked.sel(energy = current_energy)*total_intensity_pol0_corr, 97.5), 
            #                             np.nanpercentile(current_scan_pol90_unstacked.sel(energy = current_energy)*total_intensity_pol90_corr, 97.5)]))

            current_vmax_cake = current_vmax_detImage

            fig, axes = plt.subplots(2, 2, figsize = (7, 7))

            # im_pol0 = axes[0, 0].imshow(current_scan_pol0_unstacked.sel(energy = current_energy, polarization = 0).where(~np.array(integ.mask)),
            im_pol0 = axes[0, 0].imshow(current_scan_pol0_unstacked.sel(energy = current_energy, polarization = 0)*total_intensity_pol0_corr,
                            # norm = LogNorm(1, np.nanmax(current_vmax_detImage)),
                            vmin = 0,
                            vmax = np.nanmax(current_vmax_detImage),
                            cmap = cmap,
                            aspect = 1)

            axes[0, 0].set_xticks([])
            axes[0, 0].set_yticks([])

            # im_pol90 = axes[0, 1].imshow(current_scan_pol90_unstacked.sel(energy = current_energy, polarization = 90).where(~np.array(integ.mask)),
            im_pol90 = axes[0, 1].imshow(current_scan_pol90_unstacked.sel(energy = current_energy, polarization = 90)*total_intensity_pol90_corr,
                            # norm = LogNorm(1, np.nanmax(current_vmax_detImage)),
                            vmin = 0,
                            vmax = np.nanmax(current_vmax_detImage),
                            cmap = cmap,
                            aspect = 1)

            axes[0, 1].set_xticks([])
            axes[0, 1].set_yticks([])

            cake_pol0 = integrated_data_pol0_unstacked.sel(energy = current_energy, q = slice(start_q, end_q))*total_intensity_pol0_corr
            
            im = cake_pol0.plot(
                x = 'chi', y = 'q',
                ax = axes[1, 0],
                # norm = LogNorm(1, current_vmax_detImage[-1]),
                vmin = 0,
                vmax = np.nanmax(current_vmax_detImage),
                cmap = cmap,
                add_colorbar = False)

            axes[1, 0].set_ylabel('$\it{q}$ (Å$^{-1}$)')
            axes[1, 0].set_xlabel('Azimuth, $\it{χ}$')
            axes[1, 0].set_yticks([0.01, 0.03, 0.05, 0.07, 0.09])
            axes[1, 0].set_xticks([-180, -90, 0, 90, 180], ['', '-90°', '0°', '90°', ''])
            # axes[1, 0].set_xticks([-180, -90, 0, 90, 180], ['-π', '', '0', '', 'π'])

            axes[1, 0].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
            axes[1, 0].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
            axes[1, 0].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', right = 'on')
            axes[1, 0].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', right = 'on')
            
            
            cake_pol90 = integrated_data_pol90_unstacked.sel(energy = current_energy, q = slice(start_q, end_q))*total_intensity_pol90_corr

            im = cake_pol90.plot(
                x = 'chi', y = 'q',
                ax = axes[1, 1],
                # norm = LogNorm(1, current_vmax_detImage[-1]),
                vmin = 0,
                vmax = np.nanmax(current_vmax_detImage),
                cmap = cmap,
                add_colorbar = False)

            axes[1, 1].set_xlabel('Azimuth, $\it{χ}$')
            axes[1, 1].set_ylabel('')
            axes[1, 1].set_xticks([-180, -90, 0, 90, 180], ['', '-90°', '0°', '90°', '180°'])
            # axes[1, 1].set_xticks([-180, -90, 0, 90, 180], ['-π', '', '0', '', 'π'])
            axes[1, 1].set_yticks([0.01, 0.03, 0.05, 0.07, 0.09])
            axes[1, 1].set_yticklabels([])

            axes[1, 1].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
            axes[1, 1].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
            axes[1, 1].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', right = 'on')
            axes[1, 1].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', right = 'on')

            cbar_ax = fig.add_axes([0.975, 0.145, 0.02, 0.725])
            cbar = fig.colorbar(im_pol90, cax = cbar_ax)
            # cbar.ax.set_ylabel('Intensity (a.u.)')

            fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 17.5, y = 0.95, x = 0.55)
            axes[0, 0].set_title('')
            axes[0, 1].set_title('')
            axes[1, 0].set_title('')
            axes[1, 1].set_title('')

            axes[0, 0].text(0.975, 0.91, f'{current_energy:.2f} eV, 0°', color = 'white', fontsize = 15, transform = axes[0, 0].transAxes, horizontalalignment = 'right')
            axes[0, 1].text(0.975, 0.91, f'{current_energy:.2f} eV, 90°', color = 'white', fontsize = 15, transform = axes[0, 1].transAxes, horizontalalignment = 'right')
            axes[1, 0].text(0.975, 0.91, f'{current_energy:.2f} eV, 0°', color = 'white', fontsize = 15, transform = axes[1, 0].transAxes, horizontalalignment = 'right')
            axes[1, 1].text(0.975, 0.91, f'{current_energy:.2f} eV, 90°', color = 'white', fontsize = 15, transform = axes[1, 1].transAxes, horizontalalignment = 'right')

            plt.tight_layout()

            # plt.show()

            save_fig(fig, current_scan_pol0, current_energy, abs_detImage_save_dir, 'detImage', 450)





            [fig, axes] = plt.subplots(2, 2, figsize = (7, 7))
            fig.suptitle(f'{current_scan_pol0.sample_name}, {current_edge} edge', fontsize = 20, y = 0.975, x = 0.5)
            axes[1, 1].axis('off')

            pol0_para = integrated_data_pol0_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = current_energy, q = slice(start_q, end_q,), polarization = 0)*total_intensity_pol0_corr
            pol0_perp = integrated_data_pol0_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = current_energy, q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr

            pol90_para = integrated_data_pol90_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = current_energy, q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr
            pol90_perp = integrated_data_pol90_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = current_energy, q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr

            para = pol0_para + pol90_para
            perp = pol0_perp + pol90_perp

            axes[0, 0].errorbar(np.asarray(para.q).flatten(), np.asarray(para).flatten(), np.sqrt(para.values).flatten(),
                        linestyle ='none', color = (57/255, 115/255, 180/255), alpha=0.25)

            axes[0, 0].plot(para.q, para, '-', color = (57/255, 115/255, 180/255), label = 'Parallel')

            axes[0, 0].errorbar(np.asarray(perp.q).flatten(), np.asarray(perp).flatten(), np.sqrt(perp.values).flatten(),
                        linestyle ='none', color = (170/255, 21/255, 40/255), alpha=0.25)

            axes[0, 0].plot(perp.q, perp, '-', color = (170/255, 21/255, 40/255), label = 'Perpendicular')

            axes[0, 0].legend(loc = 'upper right', frameon = False, fontsize = 15)

            axes[0, 0].set_title('')
            axes[0, 0].text(0.395, 0.6, f'{current_energy:.2f} eV', color = 'black', fontsize = 15, transform = axes[0, 0].transAxes, horizontalalignment = 'left')

            axes[0, 0].set_ylabel('Intensity (a.u.)')
            axes[0, 0].set_xlabel('')
            axes[0, 0].set_xlim([start_q, end_q])
            axes[0, 0].set_ylim([np.nanmin(current_linecuts_ylim_min), 
                                 50*(np.nanmax(current_linecuts_ylim_max) - np.nanmin(current_linecuts_ylim_min)) + np.nanmin(current_linecuts_ylim_min)])
            axes[0, 0].set_yticklabels([])
            axes[0, 0].set_xticklabels([])
            axes[0, 0].set_xticks([0.01, 0.03, 0.05, 0.07, 0.09])

            axes[0, 0].yaxis.set_major_formatter(ticker.FuncFormatter(fmt))
            axes[0, 0].set_yscale('log')

            axes[0, 0].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', top = 'on')
            axes[0, 0].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', top = 'on')
            axes[0, 0].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
            axes[0, 0].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

            # current_linecuts_ylim_min.append(np.nanmin(axes[0, 0].get_ylim()))
            # current_linecuts_ylim_max.append(np.nanmax(axes[0, 0].get_ylim()))


            FY_NEXAFS.plot(
                ax = axes[0, 1],
                color = 'k',
                alpha = 0.2)

            integrated_data_unstacked.mean('chi').sel(q = slice(0.055, 0.075), energy = slice(start_en, current_energy)).mean('q').plot(
                ax = axes[0, 1],
                color = 'k',
                alpha = 1)

            axes[0, 1].plot(current_energy, integrated_data_unstacked.mean('chi').sel(q = slice(0.055, 0.075)).mean('q').sel(energy = current_energy), 'o', color = 'black')

            axes[0, 1].set_title('Fluorescence NEXAFS', color = 'black', fontsize = 18)
            # axes[0, 1].text(0.975, 0.05, f'FY NEXAFS', color = 'black', fontsize = 15, transform = axes[0, 1].transAxes, horizontalalignment = 'right')

            # axes[0, 1].set_ylabel('Intensity (a.u.)')
            axes[0, 1].set_xlabel('Energy (eV)')
            axes[0, 1].set_xlim([start_en, end_en])
            axes[0, 1].set_ylim([0.9*np.nanmin(FY_NEXAFS), 
                                 1.1*np.nanmax(FY_NEXAFS)])
            axes[0, 1].set_yticklabels([])
            # axes[0, 1].set_yscale('log')

            axes[0, 1].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', top = 'on')
            axes[0, 1].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', top = 'on')
            axes[0, 1].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
            axes[0, 1].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')


            pol0_para = integrated_data_pol0_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = slice(start_en, current_energy), q = slice(start_q, end_q,), polarization = 0)*total_intensity_pol0_corr
            pol0_perp = integrated_data_pol0_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = slice(start_en, current_energy), q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr

            pol90_para = integrated_data_pol90_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = slice(start_en, current_energy), q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr
            pol90_perp = integrated_data_pol90_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = slice(start_en, current_energy), q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr

            AR_pol0 = (pol0_para - pol0_perp)/(pol0_para + pol0_para)
            AR_pol90 = (pol90_para - pol90_perp)/(pol90_para + pol90_para)

            AR = (AR_pol0 + AR_pol90)/2
            
            if len(AR.energy) < 2:
                pol0_para = integrated_data_pol0_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = slice(start_en, end_en), q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr
                pol0_perp = integrated_data_pol0_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = slice(start_en, end_en), q = slice(start_q, end_q), polarization = 0)*total_intensity_pol0_corr

                pol90_para = integrated_data_pol90_unstacked.rsoxs.slice_chi(180, chi_width = 22.5).sel(energy = slice(start_en, end_en), q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr
                pol90_perp = integrated_data_pol90_unstacked.rsoxs.slice_chi(90, chi_width = 22.5).sel(energy = slice(start_en, end_en), q = slice(start_q, end_q), polarization = 90)*total_intensity_pol90_corr

                AR_pol0 = (pol0_para - pol0_perp)/(pol0_para + pol0_para)
                AR_pol90 = (pol90_para - pol90_perp)/(pol90_para + pol90_para)

                AR1 = (AR_pol0 + AR_pol90)/2

                scale = np.nanmax([abs(np.nanpercentile(AR1.sel(energy = slice(start_en, end_en)), 1)), abs(np.nanpercentile(AR1.sel(energy = slice(start_en, end_en)), 99))])
                # scale = np.nanmax([abs(np.nanmin(AR.sel(energy = slice(280, 295)))), abs(np.nanmax(AR.sel(energy = slice(280, 295))))])

                im_cbar = AR1.plot(
                    x = 'q', y = 'energy',
                    ax = axes[1, 0],
                    vmin = -scale,
                    vmax = scale,
                    cmap = 'RdBu',
                    add_colorbar = False)

                cbar_ax = fig.add_axes([0.5, 0.14, 0.02, 0.31])
                cbar = fig.colorbar(im_cbar, cax = cbar_ax)

                # plt.setp(cbar.ax.get_yticklabels(), ha = "right")
                # cbar.ax.tick_params(pad = 50)
                # cbar.ax.set_ylabel('Anisotropy')

                axes[1, 0].set_title('')
                # axes[1, 0].text(0.975, 0.05, f'AR', color = 'black', fontsize = 15, transform = axes[1, 0].transAxes, horizontalalignment = 'right')

                axes[1, 0].set_xlabel('$\it{q}$ (Å$^{-1}$)')
                axes[1, 0].set_ylabel('Energy (eV)')
                axes[1, 0].set_xlim([start_q, end_q])
                axes[1, 0].set_xticks([0.01, 0.03, 0.05, 0.07, 0.09], ['', '0.03', '0.05', '0.07', '0.09'])
                axes[1, 0].set_ylim([AR1.energy[0], (AR1.energy[1] - AR1.energy[0])/2 + AR1.energy[0]])
                # axes[1, 0].set_ylim([np.nanmin(AR1.energy), np.nanmax(AR1.energy)])
                axes[1, 0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

                # ticks = [tick for tick in axes[1, 0].get_yticks()]
                # del ticks[-1]
                # axes[1, 0].set_yticks(ticks)

                axes[1, 0].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', top = 'on')
                axes[1, 0].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', top = 'on')
                axes[1, 0].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
                axes[1, 0].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

            else:

                AR1.plot(
                    x = 'q', y = 'energy',
                    ax = axes[1, 0],
                    vmin = -scale,
                    vmax = scale,
                    cmap = 'RdBu',
                    add_colorbar = False,
                    alpha = 0.2)

                im_cbar = AR.plot(
                    x = 'q', y = 'energy',
                    ax = axes[1, 0],
                    vmin = -scale,
                    vmax = scale,
                    cmap = 'RdBu',
                    add_colorbar = False)

                cbar_ax = fig.add_axes([0.5, 0.14, 0.02, 0.31])
                cbar = fig.colorbar(im_cbar, cax = cbar_ax)

                # plt.setp(cbar.ax.get_yticklabels(), ha = "right")
                # cbar.ax.tick_params(pad = 50)
                cbar.ax.set_ylabel('Anisotropy')

                axes[1, 0].set_title('')
                # axes[1, 0].text(0.975, 0.05, f'AR', color = 'black', fontsize = 15, transform = axes[1, 0].transAxes, horizontalalignment = 'right')

                axes[1, 0].set_xlabel('$\it{q}$ (Å$^{-1}$)')
                axes[1, 0].set_ylabel('Energy (eV)')
                axes[1, 0].set_xlim([start_q, end_q])
                axes[1, 0].set_xticks([0.01, 0.03, 0.05, 0.07, 0.09], ['', '0.03', '0.05', '0.07', '0.09'])

                # ticks = [tick for tick in axes[1, 0].get_yticks()]
                # del ticks[-1]
                # axes[1, 0].set_yticks(ticks)
                axes[1, 0].set_ylim([np.nanmin(AR.energy), np.nanmax(AR.energy)])
                axes[1, 0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

                axes[1, 0].xaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'inout', top = 'on')
                axes[1, 0].xaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'inout', top = 'on')
                axes[1, 0].yaxis.set_tick_params(which = 'major', size = 5, width = 2, direction = 'in', right = 'on')
                axes[1, 0].yaxis.set_tick_params(which = 'minor', size = 2.5, width = 2, direction = 'in', right = 'on')

            clear_output(wait = True)

            # plt.tight_layout()

            plt.show()
            
            save_fig(fig, current_scan_pol0, current_energy, abs_linecuts_save_dir, 'linecuts', 450)
            
            print(f'{len(scans)} scans remaining')
            
        clear_output(wait = True)    
        
        print(f'{len(scans)} scans remaining')
        
        print('Making gif of abs det images')    
        make_gif(current_scan_pol0, abs_detImage_save_dir, reduced_save_dir, 'abs_detImage')
        
        print('Making gif of abs linecuts images') 
        make_gif(current_scan_pol0, abs_linecuts_save_dir, reduced_save_dir, 'abs_linecuts')
        
        print('ans gifs done, purging temporary images')  
        shutil.rmtree(abs_detImage_save_dir)
        shutil.rmtree(abs_linecuts_save_dir)
        
        # save variables to pickle
        with open('PyHyperScattering_Batch_SST1_JupyterHub.pkl', 'wb') as file:
            pickle.dump([sample_names, edges, ARs, AR_scales, paras, para_scales, perps, perp_scales, circs, circ_scales], file, protocol = -1)

    else:
        print(f'{current_scan_pol0.sampleid} already processed, moving on')



        
        
        
        
        

print(f'{len(scans)} scans leftover')
```
